```python
import requests
import json
import pandas as pd
from fake_useragent import UserAgent
```

```python
latest_url = 'https://www.dcard.tw/_api/forums/{}/posts?popular=false&limit=100'
# article_url = 'https://www.dcard.tw/_api/posts?forum={}&limit=100'

forums = ['food', 'boba', 'cvs', 'fastfood', 'travel', 'jp_drama', 'japan_star']
```

```python
columns_need = ['createdAt', 'categories', 'title', 'excerpt', 'topics', 'commentCount', 'likeCount', 'id', 'school', 'gender']
columns_name = ['發文時間', '分類', '標題', '摘要', '標籤', '留言數', '愛心數', 'ID', '發文者', '性別']
```

```python
def get_df_from_api(api_url, forum):
    ua = UserAgent()
    headers = {'user-agent': ua.random}
    url = api_url.format(forum)
    resp = requests.get(url, headers=headers)
    if resp.status_code ==200:
        df = pd.DataFrame(json.loads(resp.text))
        df = df.loc[:, columns_need]
        df.columns = columns_name
        df['發文時間'] = df['發文時間'].apply(lambda x: x[:10])
        df['標籤'] = df['標籤'].apply(lambda x: ' / '.join(x))
    return df
```

```python
all_df = {}
for forum in forums:
    all_df[forum] = get_df_from_api(latest_url, forum)
```
